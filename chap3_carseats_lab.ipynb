```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
import statsmodels.api as sm
```

```python
flc = '/Users/pinqingkan/Desktop/Codes/Course_ISLR/data/'
fname = flc + 'Carseats.csv'
X = pd.read_csv(fname, index_col = 0)
```

```python
X.describe(include = 'all')
```

Encode qualitative features

```python
ShelveLoc2 = pd.get_dummies(X.ShelveLoc)
fcols2 = []
for i in ShelveLoc2.columns:
    fcols2.append('ShelveLoc_' + i)
ShelveLoc2.columns = fcols2
Urban2 = pd.get_dummies(X.Urban)
fcols2 = []
for i in Urban2.columns:
    fcols2.append('Urban_' + i)
Urban2.columns = fcols2 
US2 = pd.get_dummies(X.US)
fcols2 = []
for i in US2.columns:
    fcols2.append('US_' + i)
US2.columns = fcols2 
```

```python
# replace qualitative features dummy-coded values
X2 = pd.concat([X.drop(['ShelveLoc', 'Urban', 'US'], axis = 1), ShelveLoc2.iloc[:, :-1], Urban2.iloc[:, :-1], US2.iloc[:, :-1]], axis = 1)
```

```python
# add interaction terms
xInter = pd.concat([X.Income * X.Advertising, X.Price * X.Age], axis = 1)
xInter.columns = ['Income*Advertising', 'Price*Age']
# multiple linear regression
xx = pd.concat([X2.drop(['Sales'], axis = 1), xInter], axis = 1)
xx = sm.add_constant(xx) # OLS does not include intercept by default
yy = X.Sales.copy()
results0 = sm.OLS(yy, xx).fit()
```

```python
results0.summary()
```
